```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
info = pd.read_csv('Cars Data.csv') 
info
```

```python
#no-1) Find the column which has max null values in the dataset then fill it with mean of that column.
info.isnull().sum()

```

```python
info['Cylinders'].fillna(info['Cylinders'].mean(),inplace=True)
info
```

```python
#no-2)Check what are the different types of Make are there in our dataset.And what is the count of each make in the dataset.
info['Make'].value_counts()
```

```python
#no-3)show all the records where origin is asia or europe.
info.loc[(info['Origin']=='Asia')|(info['Origin']=='Europe')]
          
```

```python
#no-3)remove all the records where weight is above 4000.
info[~(info['Weight']>4000)]
```

```python
#no-5)increase all the values of 'MPG_City'column by 3.
info['MPG_City'] = info['MPG_City'].apply(lambda x:x+3)
info['MPG_City']
```

```python
#plot a scatter plot for weight vs wheel base.
x= info['Weight'].tolist()
y= info['Wheelbase'].tolist()
plt.scatter(x,y,label='skitscat',color='r')
plt.xlabel('Weight')
plt.ylabel('Wheelbase')
plt.legend()
plt.show()
```
